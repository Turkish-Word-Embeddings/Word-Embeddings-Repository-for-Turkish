```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a dataframe in this format
# algorithm  syntactic-verb-MRR  syntactic-noun-MRR semantic-MRR 

analogy = [{'Syntactic Noun MRR': 0.318, 'Syntactic Verb MRR': 0.604, 'Semantic MRR': 0.535},  # word2vec 5 epoch
        {'Syntactic Noun MRR': 0.329, 'Syntactic Verb MRR': 0.609, 'Semantic MRR': 0.551},     # word2vec 10 epoch
        {'Syntactic Noun MRR': 0.360, 'Syntactic Verb MRR': 0.718, 'Semantic MRR': 0.258},     # fasttext 5 epoch
        {'Syntactic Noun MRR': 0.371, 'Syntactic Verb MRR': 0.697, 'Semantic MRR': 0.303},     # fasttext 10 epoch
        {'Syntactic Noun MRR': 0.514, 'Syntactic Verb MRR': 0.801, 'Semantic MRR': 0.346}]     # fasttext Facebook AI 

# algorithm Syntactic-Similarity-Pearson  Syntactic-Similarity-Spearman  Semantic-Similarity-Pearson Semantic-Similarity-Spearman

similarity = [{'Syntactic Similarity Pearson': 83.80, 'Syntactic Similarity Spearman': 82.57, 'Semantic Similarity Pearson': 70.72, 'Semantic Similarity Spearman': 74.33}, # word2vec 5 epoch
        {'Syntactic Similarity Pearson': 82.53, 'Syntactic Similarity Spearman': 81.68, 'Semantic Similarity Pearson': 70.77, 'Semantic Similarity Spearman': 74.57},       # word2vec 10 epoch
        {'Syntactic Similarity Pearson': 61.12, 'Syntactic Similarity Spearman': 62.55, 'Semantic Similarity Pearson': 67.45 , 'Semantic Similarity Spearman': 70.37},       # fasttext 5 epoch
        {'Syntactic Similarity Pearson': 66.51, 'Syntactic Similarity Spearman': 68.16, 'Semantic Similarity Pearson':67.35 , 'Semantic Similarity Spearman': 70.53},        # fasttext 10 epoch
         {'Syntactic Similarity Pearson': 76.92, 'Syntactic Similarity Spearman': 80.28, 'Semantic Similarity Pearson': 65.12, 'Semantic Similarity Spearman': 68.92}]       # fasttext facebook AI 
```

```python
analogydf = pd.DataFrame(analogy, index=['Word2Vec 5 epoch', 'Word2Vec 10 epoch', 'FastText 5 epoch', 'FastText 10 epoch', 'FastText Turkish Wordvectors'])
similaritydf = pd.DataFrame(similarity, index=['Word2Vec 5 epoch', 'Word2Vec 10 epoch', 'FastText 5 epoch', 'FastText 10 epoch', 'FastText Turkish Wordvectors'])
```

```python
# Create a list of MRR types and their positions on the x-axis
plt.figure(figsize=(15, 10), dpi=80)
plt.plot(analogydf.loc['Word2Vec 5 epoch'], label='Word2Vec 5 epoch (300 emb)')
plt.plot(analogydf.loc['Word2Vec 10 epoch'], label='Word2Vec 10 epoch (300 emb)')
plt.plot(analogydf.loc['FastText 5 epoch'], label='FastText 5 epoch (300 emb)')
plt.plot(analogydf.loc['FastText 10 epoch'], label='FastText 10 epoch (300 emb)')
plt.plot(analogydf.loc['FastText Turkish Wordvectors'], label='FastText Turkish Wordvectors (Bojanowski et. al.)')
plt.legend()
plt.xlabel('MRR Type')
plt.ylabel('MRR Value')
plt.show()
```

```python
# Bar plot
ax = analogydf.plot(kind='bar')

# Add labels, title, and legend
ax.set_ylabel('MRR Value')
ax.set_xlabel('Algorithm')
ax.legend(title='Category', bbox_to_anchor=(1, 1))
ax.set_title('MRR Comparison')

plt.show()
```

```python
# Create a list of MRR types and their positions on the x-axis
plt.figure(figsize=(15, 10), dpi=80)
plt.plot(similaritydf.loc['Word2Vec 5 epoch'], label='Word2Vec 5 epoch (300 emb)')
plt.plot(similaritydf.loc['Word2Vec 10 epoch'], label='Word2Vec 10 epoch (300 emb)')
plt.plot(similaritydf.loc['FastText 5 epoch'], label='FastText 5 epoch (300 emb)')
plt.plot(similaritydf.loc['FastText 10 epoch'], label='FastText 10 epoch (300 emb)')
plt.plot(similaritydf.loc['FastText Turkish Wordvectors'], label='FastText Turkish Wordvectors (Bojanowski et. al.)')
plt.legend()
plt.xlabel('MRR Type')
plt.ylabel('MRR Value')
plt.show()
```

```python
# Bar plot
ax = similaritydf.plot(kind='bar')

# Add labels, title, and legend
ax.set_ylabel('MRR Value')
ax.set_xlabel('Algorithm')
ax.legend(title='Statistics', bbox_to_anchor=(1, 1))
ax.set_title('Pearson/Spearman Comparison')

plt.show()
```
