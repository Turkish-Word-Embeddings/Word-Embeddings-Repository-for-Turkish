```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a dataframe in this format
# algorithm  syntactic-verb-MRR  syntactic-noun-MRR semantic-MRR 

analogy = [{'Syntactic Noun MRR': 0.318, 'Syntactic Verb MRR': 0.604, 'Semantic MRR': 0.535},
        {'Syntactic Noun MRR': 0.329, 'Syntactic Verb MRR': 0.609, 'Semantic MRR': 0.551},
        {'Syntactic Noun MRR': 0.360, 'Syntactic Verb MRR': 0.718, 'Semantic MRR': 0.258}]

# algorithm Syntactic-Similarity-Pearson  Syntactic-Similarity-Spearman  Semantic-Similarity-Pearson Semantic-Similarity-Spearman

similarity = [{'Syntactic Similarity Pearson': 83.80, 'Syntactic Similarity Spearman': 82.57, 'Semantic Similarity Pearson': 70.72, 'Semantic Similarity Spearman': 74.33},
        {'Syntactic Similarity Pearson': 82.53, 'Syntactic Similarity Spearman': 81.68, 'Semantic Similarity Pearson': 70.77, 'Semantic Similarity Spearman': 74.57},
        {'Syntactic Similarity Pearson': 61.12, 'Syntactic Similarity Spearman': 62.55, 'Semantic Similarity Pearson':67.45 , 'Semantic Similarity Spearman': 70.37}]
```

```python
analogydf = pd.DataFrame(analogy, index=['Word2Vec 5 epoch', 'Word2Vec 10 epoch', 'FastText 10 epoch'])
similaritydf = pd.DataFrame(similarity, index=['Word2Vec 5 epoch', 'Word2Vec 10 epoch', 'FastText 10 epoch'])
```

```python
# Create a list of MRR types and their positions on the x-axis
plt.figure()
plt.plot(analogydf.loc['Word2Vec 5 epoch'], label='Word2Vec 5 epoch')
plt.plot(analogydf.loc['Word2Vec 10 epoch'], label='Word2Vec 10 epoch')
plt.plot(analogydf.loc['FastText 10 epoch'], label='FastText 10 epoch')
plt.legend()
plt.xlabel('MRR Type')
plt.ylabel('MRR Value')
plt.show()
```

```python
# Create a list of MRR types and their positions on the x-axis
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(similaritydf.loc['Word2Vec 5 epoch'], label='Word2Vec 5 epoch')
plt.plot(similaritydf.loc['Word2Vec 10 epoch'], label='Word2Vec 10 epoch')
plt.plot(similaritydf.loc['FastText 10 epoch'], label='FastText 10 epoch')
plt.legend()
plt.xlabel('MRR Type')
plt.ylabel('MRR Value')
plt.show()
```
